Histogram for Frequency

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Solar_flare_RHESSI_2004_05.csv')

df['date'] = pd.to_datetime(df[['year', 'month']].assign(DAY=1))

months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['month'] = pd.Categorical(df['date'].dt.month_name(), categories=months_order, ordered=True)

plt.figure(figsize=(12, 6))
colors = ['skyblue', 'salmon']
years_order = sorted(df['date'].dt.year.unique())

for i, year in enumerate(years_order):
    data_year = df[df['date'].dt.year == year]
    data_year['month'].value_counts().sort_index().plot(kind='bar', label=str(year), alpha=0.7, color=colors[i])

plt.title('Histogram of Months')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.legend(title='Year')
plt.show()
```
